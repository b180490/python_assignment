```python
plt.scatter(df['Price'],df['HP'])

plt.xlabel('price(rs.)')
plt.ylabel('hourse power')

plt.legend()

plt.title('Price v/s HP')
```

```python
plt.bar(df['Price'],df['HP'])

plt.xlabel('price(rs.)')
plt.ylabel('hourse power')

plt.legend()

plt.title('Price v/s HP')
```

```python
plt.hist(df['Price'])

plt.xlabel('price(rs.)')
plt.ylabel('price(rs.)')

plt.legend()

plt.title('Price v/s Price')
```
