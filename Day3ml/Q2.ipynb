Perform K-Means Clustering for the crime_std data and identify the number of clusters formed and draw inferences.

Data Description:
Murder -- Muder rates in different places of United States
Assualt- Assualt rate in different places of United States
UrbanPop - urban population in different places of United States
Rape - Rape rate in different places of United States

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("crime_std.csv")
df
```

```python
df.drop(["Unnamed: 0"],axis = 1, inplace = True)
df
```

```python
wss = []

for k in range(1,11):
    kmeans_model = KMeans(n_clusters=k, random_state=12345)

    kmeans_model.fit(df)

    wss.append(kmeans_model.inertia_)

print(wss)
    
```

```python
x = np.arange(1,11)

plt.plot(x, wss)
plt.title("Elbow method")
plt.xlabel("no of clusters")
plt.ylabel("Sum of Squares of within cluster variations")
plt.show()
```
