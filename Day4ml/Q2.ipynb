Perform Hierarchical Clustering and DBSCAN Clustering for the crime_std data and identify the number of clusters formed and draw inferences.

Data Description:
Murder -- Muder rates in different places of United States

UrbanPop - urban population in different places of United States

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster import hierarchy
from sklearn.cluster import DBSCAN
```

```python
df = pd.read_csv("crime_std.csv")
df
```

```python
df.describe()
```

```python
df.info()
```

```python
df.drop(["Unnamed: 0"],axis=1,inplace=True)
```

```python
df.columns
```

```python
model = DBSCAN(eps=4.5, min_samples=3)
model.fit(df)
```

```python
h_model = hierarchy.linkage(df,'single')
plt.figure()
result = hierarchy.dendrogram(h_model)
```

```python
h_model = hierarchy.linkage(df,'complete')
plt.figure()
result = hierarchy.dendrogram(h_model)
```

```python
h_model = hierarchy.linkage(df,'average')
plt.figure
result = hierarchy.dendrogram(h_model)
```
