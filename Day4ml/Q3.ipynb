Perform Hierarchical Clustering for the Universities_std data and identify the number of clusters formed.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster import hierarchy
```

```python
df = pd.read_csv("Universities_std.csv")
df
```

```python
df.describe()
```

```python
df.info()
```

```python
h_model = hierarchy.linkage(df,'single')
plt.figure()
result = hierarchy.dendrogram(h_model)
```

```python
h_model = hierarchy.linkage(df,'complete')
plt.figure()
result = hierarchy.dendrogram(h_model)
```
