Perform DBSCAN Clustering for the Grocery_std data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster import hierarchy
from sklearn.cluster import DBSCAN
```

```python
df = pd.read_csv("Grocery_std.csv")
df
```

```python
df.describe()
```

```python
df.info()
```

```python
h_model = hierarchy.linkage(df,'single')
plt.figure()
result = hierarchy.dendrogram(h_model)
```

```python
h_model = hierarchy.linkage(df,'complete')
plt.figure()
result = hierarchy.dendrogram(h_model)
```

```python
h_model = hierarchy.linkage(df,'average')
plt.figure()
result = hierarchy.dendrogram(h_model)
```

```python
model = DBSCAN(eps=3.50, min_samples=7)
model.fit(df)
```
