```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("wine_std.csv")
df
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df.columns
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=1)

pca.fit(df)

new_component = pca.transform(df)

df['pc'] = new_component

df
```
