```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Zoo.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# create an encoder object
encoder = LabelEncoder()

# fit the data
encoder.fit(df['type'])


# transform the Gender column
df['type'] = encoder.transform(df['type'])
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.drop(['airborne','predator','fins'],axis=1,inplace=True)
```

```python
X = df.drop(['type'],axis=1)
Y = df['type']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X,Y,train_size=0.7,random_state=12345)
```

```python
def build_dt_model():
    from sklearn.tree import DecisionTreeClassifier
    
    # create a model
    model = DecisionTreeClassifier()

    # train the model
    model.fit(x_train, y_train)

    return model
```

```python
def build_rf_model():
    from sklearn.ensemble import RandomForestClassifier

    # create a model
    model = RandomForestClassifier()

    # train the model
    model.fit(x_train, y_train)

    return model
```

```python
def build_cb_model():
    from catboost import CatBoostClassifier
    
    # create a model
    model = CatBoostClassifier()
    
    # train the model
    model.fit(x_train, y_train)
    
    return model
```

```python
def build_xgb_model():
    from xgboost import XGBClassifier
    
    # create a model
    model = XGBClassifier()
    
    # train the model
    model.fit(x_train, y_train)
    
    return model
```

```python
def evaluate_model(model):
    y_true = y_test
    y_pred = model.predict(x_test)
    from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

    accuracy = f"{accuracy_score(y_true, y_pred):.2f}"
    precision = f"{precision_score(y_true, y_pred,average='macro'):.2f}"
    recall = f"{recall_score(y_true, y_pred,average='macro'):.2f}"
    f1 = f"{f1_score(y_true, y_pred,average='macro'):.2f}"

    return accuracy, precision, recall, f1
```

```python
evaluation_report = []

model_dt = build_dt_model()
evaluation_report.append({"model": "Decision Tree", "metrics": evaluate_model(model_dt)})

model_rf = build_rf_model()
evaluation_report.append({"model": "Random Forest", "metrics": evaluate_model(model_rf)})

model_cb = build_cb_model()
evaluation_report.append({"model": "CatBoost", "metrics": evaluate_model(model_cb)})

model_xgb = build_xgb_model()
evaluation_report.append({"model": "XGB", "metrics": evaluate_model(model_xgb)})
```

```python
print(f"{'Algorithm':<20} => 'Accuracy', 'precision', 'recall', 'f1'")
print()
for info in evaluation_report:
    print(f"{info['model']:<20} => {info['metrics']}")
```
